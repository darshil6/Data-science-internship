```python

# Select only numerical columns for outlier detection
numerical_cols = data.select_dtypes(include=['int64', 'float64'])

# Finding outliers using IQR method
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Check for outliers in each numerical column
outliers_dict = {}
for column in numerical_cols.columns:
    outliers = find_outliers_iqr(data, column)
    outliers_dict[column] = outliers
    print(f"Outliers in {column}:")
    print(outliers[column])
    print("-" * 40)

# Summary: Printing the number of outliers in each column
for column, outliers in outliers_dict.items():
    print(f"{column}: {len(outliers)} outliers")
```

```python
# Plotting histograms and PDFs for each numerical column
for column in numerical_cols.columns:
    plt.figure(figsize=(12, 6))

    # Histogram with KDE (Kernel Density Estimate)
    sns.histplot(data[column], kde=True, bins=30, color='blue', stat="density")

    plt.title(f'Probability and Frequency Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```

```python
# Select categorical columns for frequency distribution analysis
categorical_cols = data.select_dtypes(include=['object'])

# Set up the plotting style
sns.set_style("whitegrid")

# Plotting countplots for each categorical column
for column in categorical_cols.columns:
    plt.figure(figsize=(12, 6))
    sns.countplot(y=column, data=data, palette='viridis')

    plt.title(f'Frequency Distribution of {column}')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.show()
```

```python
# Select categorical columns for frequency distribution analysis
categorical_cols = data.select_dtypes(include=['object'])

# Plotting countplots for each categorical column (limited to first 3 for brevity)
for column in categorical_cols.columns[:3]:  # Change the slice as needed
    plt.figure(figsize=(10, 5))
    sns.countplot(y=column, data=data, palette='viridis')
    plt.title(f'Frequency Distribution of {column}')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.show()

    # Observations
    counts = data[column].value_counts()
    print(f"Observations for {column}:")

    # Check for the most common categories
    for category, count in counts.items():
        print(f" - {category}: {count} entries")

    # Specific observations based on counts
    if counts.max() > 1000:  # Example threshold for observation
        print(f" - The category '{counts.idxmax()}' has a significantly high count.")
    else:
        print(" - All categories have relatively low counts.")

    print("-" * 40)
```
