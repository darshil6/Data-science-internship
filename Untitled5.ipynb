```python
# Select two categorical columns for stacked bar plot analysis
categorical_column_1 = 'Gender'
categorical_column_2 = 'CollegeTier'

# Create a contingency table using pandas crosstab
contingency_table = pd.crosstab(data[categorical_column_1], data[categorical_column_2])

# Plotting the stacked bar plot
contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

# Customize the plot
plt.title(f'Stacked Bar Plot of {categorical_column_1} vs {categorical_column_2}')
plt.xlabel(categorical_column_1)
plt.ylabel('Count')
plt.legend(title=categorical_column_2, loc='upper right')
plt.show()

```

```python
# Filter data for relevant designations and fresh graduates (assuming 'DOJ' or similar column represents joining date)
relevant_designations = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
filtered_data = data[data['Designation'].isin(relevant_designations)]

# Consider only recent graduates - assuming GraduationYear can be used to identify fresh graduates
fresh_graduates = filtered_data[filtered_data['GraduationYear'] >= 2018]

# Check if there are any fresh graduates
if len(fresh_graduates) > 0:
    # Analyzing Salary
    salary_mean = fresh_graduates['Salary'].mean()
    salary_median = fresh_graduates['Salary'].median()

    # Filtering for salaries within the range of 2.5 to 3 lakhs
    within_range = fresh_graduates[(fresh_graduates['Salary'] >= 250000) & (fresh_graduates['Salary'] <= 300000)]

    # Display results
    print(f"Total number of relevant job titles: {len(fresh_graduates)}")
    print(f"Number of fresh graduates earning between 2.5 to 3 lakhs: {len(within_range)}")
    print(f"Percentage of fresh graduates earning between 2.5 to 3 lakhs: {(len(within_range) / len(fresh_graduates)) * 100:.2f}%")
    print(f"Mean Salary of fresh graduates: {salary_mean}")
    print(f"Median Salary of fresh graduates: {salary_median}")
else:
    print("No fresh graduates found with the given criteria.")
```

```python
# Create a contingency table for Gender and Specialization
contingency_table_gender_specialization = pd.crosstab(data['Gender'], data['Specialization'])

# Plotting the stacked bar plot
contingency_table_gender_specialization.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')

# Customize the plot
plt.title('Stacked Bar Plot of Gender vs Specialization')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Specialization', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.show()

# Observations
print("Observations for the relationship between Gender and Specialization:")
for gender in contingency_table_gender_specialization.index:
    print(f" - {gender}:")
    for specialization in contingency_table_gender_specialization.columns:
        count = contingency_table_gender_specialization.loc[gender, specialization]
        if count > 0:
            print(f"    - {specialization}: {count} entries")
print("-" * 40)

```
